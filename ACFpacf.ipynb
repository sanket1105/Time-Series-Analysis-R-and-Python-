##  Simulate and plot the autocorrelation functions for an AR(1) process with ùúô1 = 0.3:

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


# Define AR(1) model parameters
phi = 0.3
ar = np.array([1, -phi])

# Simulate 150 observations
np.random.seed(123)
ar_process = ArmaProcess(ar)
ar_simulated = ar_process.generate_sample(150)

# Plot AR(1) time series and sample ACF
plt.plot(ar_simulated)
plt.title('AR(1) Time Series')

plot_acf(ar_simulated)
plt.title('Sample Autocorrelation Function')
plt.show()
```

## Simulate and plot the autocorrelation functions for an MA(2) process with ùúÉ1 = ‚àí1 and ùúÉ2 = ‚àí0.6:

```python

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


# Define MA(2) model parameters
theta = np.array([1, -1, -0.6])
ma = np.array([1])

# Simulate 150 observations
np.random.seed(123)
ma_process = ArmaProcess(ma, theta)
ma_simulated = ma_process.generate_sample(150)

# Plot MA(2) time series and sample ACF
plt.plot(ma_simulated)
plt.title('MA(2) Time Series')

plot_acf(ma_simulated)
plt.title('Sample Autocorrelation Function')
plt.show()

```
